```python
#Programa que resuleva a la ecuacion de calor de la imagen.
print("--- Problema: Ecuación de Calor ---")
print("Ecuación: du/dt = d^2u/dx^2")
print("Dominio: 0 < x < 1, 0 <= t")
print("\nCondiciones:")
print("  - Frontera: u(0, t) = 0, u(1, t) = 0")
print("  - Inicial:  u(x, 0) = sin(pi*x)")
print("-----------------------------------------------------")
```

```python
import numpy as np
import matplotlib.pyplot as plt

#definimos los parametros
L=1
t_fin=0.5

#definimos ecuacion
def ex_solu(x, t):
  return np.sin(np.pi*x/L)*np.exp(-t*np.pi**2/L**2)


def solucion_ecuacion(h, k, t_fin, L):
  r=k/(h**2)

  if r>0.5:
    print("CASO INESTABLE, ya que r es mayor a 0.5, donde su valor de r es: ", r)
  else:
    print("CASO ESTABLE, ya que r es menor a 0.5, donde su valor de r es: ", r)

  #ahora, debemos definir los intervalos:
  num_inter=int(L/h)
  coords=np.linspace(0, L, num_inter)

  num_tiempo=int(t_fin/k)
  u_j=np.sin(np.pi*coords)

  for j in range(num_tiempo):
    u_j_plus_1 = np.copy(u_j)
    u_j_plus_1[1:-1] = u_j[1:-1] + r * (u_j[2:] - 2 * u_j[1:-1] + u_j[:-2])
    u_j = u_j_plus_1
  return u_j, r, coords

#CASO A
h_val = 0.1
k_val = 0.0005
solunum, r_val, coordenadas = solucion_ecuacion(h_val, k_val, t_fin, L)

print("----------------------")
print("CASO A")
print("Solucion de T =", t_fin)
print("----------------------")
print(solunum)
print("----------------------")
print("Valor de r:", r_val)
print("----------------------")
print("Coordenadas:", coordenadas)
print()
print()

#CASO B
h_val = 0.1
k_val = 0.01
solunum, r_val, coordenadas = solucion_ecuacion(h_val, k_val, t_fin, L)

print("----------------------")
print("CASO B")
print("Solucion de T =", t_fin)
print("----------------------")
print(solunum)
print("----------------------")
print("Valor de r:", r_val)
print("----------------------")
```
